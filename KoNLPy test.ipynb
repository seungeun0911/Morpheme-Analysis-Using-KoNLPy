### 추출된 키워드 빈도수 확인

추출된 명사들을 카운트하기 위해 'collections'의 'Counter'를 사용했습니다. 
'.most_common'을 이용하면 빈도수가 높은 n개의 명사만 추출할 수 있습니다.
100개를 이용하였고, 그 중에서 조건(2글자 이상 49글자 미만)에 맞는 단어 50개만인 'for'문을 이용하여 [tag_count] 리스트에 'append'를 이용하여 추가하였습니다.
이후 키워드만 활용하기 위하여 단어만 별도의 리스트에 따로 'append'했습니다

```python
from collections import Counter
count = Counter(hannanum.nouns(text))

tag_count = []
tags = []


for n, c in count.most_common(100):
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 49:
        tag_count.append(dics)
        tags.append(dics['tag'])
```

```python
for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
```
