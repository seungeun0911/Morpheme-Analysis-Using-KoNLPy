## Daum 뉴스 크롤링

```python
import requests # 원격 서버 data 요청
from bs4 import BeautifulSoup # html parser

url = "http://media.daum.net"
```

```python
# 1. url 요청
req = requests.get(url)
print(req) # 
src = req.text # text형식으로
```

```python
# 2. html 파싱
html = BeautifulSoup(src, "html.parser")
html
```

```python
# 3. tag[속성='값']
links = html.select("a[class='link_txt']")
len(links) # 115

links[:10]
links[-10:]
```

```python
# 4. tag 내용 수집
crawling_data = []

for a in links :
    #print(a.string)
    cont = a.string
    if cont : # None 제외
        cont = str(cont)
        # strip() : 문장 끝 불용어 제거
        crawling_data.append(cont.strip())
        
crawling_data
```

```python
#Dataframe으로 만들기
import pandas as pd
df = pd.DataFrame({'내용':crawling_data})

#df 저장
df.to_csv('crawling.txt',encoding='utf-8',index=False)
```

## 형태소 분석

```python
file = open("crawling.txt", mode='r', encoding='utf-8')
doc = file.read()
file.close()
doc
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
ex_sent = kkma.sentences(doc)
ex_sent
```

```python
ex_nouns = kkma.nouns(doc)
ex_nouns
```

```python
from re import match  # 전처리 위해서 정규표현식 관련 re 패키지 import
nouns = []

for sent in ex_sent : 
    for noun in kkma.nouns(sent) :
        # 단어 전처리 : 2음절 이상, 수사 제외
        if len(str(noun)) >= 2 and not(match('^[0-9]', noun)) :
            nouns.append(noun)
            
nouns
```

```python
word_count = {} # 빈 set
```

```python
for noun in nouns :
    word_count[noun] = word_count.get(noun, 0) + 1

word_count
```

```python
from collections import Counter

counter = Counter(word_count)

top5 = counter.most_common(5)
top5
```

## 데이터 시각화

```python
# data -> nouns 추출

kkma = Kkma() 

ex_nouns = [] # 단어 저장

for sent in crawling_data : # doc -> sentence
    for noun in kkma.nouns(sent) : # sentence -> noun
        ex_nouns.append(noun)

ex_nouns
len(ex_nouns)
```

```python
# 명사 전처리 : 2음절 이상, 숫자 제외 -> 단어사전(dict)
from re import match

nouns_count = {} # 빈 set

for noun in ex_nouns :
    # 2음절 이상, 숫자 제외
    if len(str(noun)) > 1 and not(match('^[0-9]', noun)) :
        nouns_count[noun] = nouns_count.get(noun, 0) + 1

nouns_count
```

```python
# 단어 카운트
from collections import Counter

counter = Counter(nouns_count)
top15 = counter.most_common(n=15)
top15
```

```python
# word cloud
import pytagcloud
import webbrowser

# tag에 color, size, tag 사전 구성 
word_count_list = pytagcloud.make_tags(top15, maxsize=100)
# maxsize : 최대 글자크기
print(word_count_list)

pytagcloud.create_tag_image(word_count_list,
                            'newsCrawling.jpg', 
                            size=(900, 600), 
                            fontname='Noto Sans CJK', rectangular=False)
```
